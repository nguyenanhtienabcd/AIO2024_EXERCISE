```python
import cv2
from ultralytics import YOLO
from ultralytics import solutions
```

```python
# khởi tạo video và định nghĩa vùng đếm
cap = cv2.VideoCapture("/content/highway.mp4")
assert cap.isOpened(), "Error reading video files"
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))
```

Vùng được xác định bởi bốn đỉnh này tạo thành một hình chữ nhật trên khung hình video. Khi xử lý video (ví dụ trong các ứng dụng đếm xe hoặc đếm người), vùng này được dùng để xác định khu vực quan tâm (Region of Interest - ROI). Các đối tượng di chuyển qua vùng này sẽ được tính hoặc xử lý theo yêu cầu của ứng dụng

```python
# khởi tạo video và định nghĩa vùng đệm
regions_points = [
    (430, 700),
    (1600, 700),
    (1600, 1080),
    (430, 1080)
]
# For rectangle region counting : top left , top right , bottom right , bottom left
```

```python
# thiết lập VideoWriter để lưu video kết quả lại
video_path = "/content/highway.mp4"
video_name = video_path.split('/')[-1].split(".")[0]
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
video_writer = cv2.VideoWriter(f"/content/{video_name}_counted.mp4", cv2.VideoWriter_fourcc(*"mp4v"), fps, (w, h))
```

```python
# khởi tạo object_counter
counter = solutions.ObjectCounter(
    show = False,
    region = regions_points,
    model = "yolo11x.pt"
)
```

```python
# process video
while cap.isOpened():
  success, imO = cap.read()
  if not success:
    print(" Video frame is empty or video provessing has been successfully completed.")
    break
  imO = counter.count(imO)
  video_writer.write(imO)

cap . release ()
video_writer.release ()
cv2.destroyAllWindows ()
```
