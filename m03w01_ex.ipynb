### 6. Resampling

```python
times_sample = pd.to_datetime(['2014-08-01', '2014-08-15', '2014-08-31'])
```

```python
# 1. Specify the data columns we want to include (i.e. exclude Year, Month, Weekday Name)
data_columns = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']

# 2. Resample to weekly frequency, aggregating with mean
dopsd_weekly_mean = dopsd_daily[data_columns].resample('W').mean()
dopsd_weekly_mean.head()
```

```python
dopsd_weekly_max = dopsd_daily[data_columns].resample('W').max()
dopsd_weekly_max.head()
```

```python
dopsd_weekly_agg = dopsd_daily[data_columns].resample('W').agg(['mean', 'sum', 'min'])
dopsd_weekly_agg.head()
```

```python
print( dopsd_daily.shape [0])
print( dopsd_weekly_mean.shape [0])
```

```python
# start and end of the date range to extract
start, end = '2015-01', '2015-06'

# plot daily and weekly resample time series together
fig, ax = plt.subplots()

#plot with original value
ax.plot(dopsd_daily.loc[start:end, 'Solar'],
        marker = '.', linestyle = '-', linewidth = 0.5, label = 'Daily')

# plot with
ax.plot(dopsd_weekly_mean.loc[start:end, 'Solar'],
        marker = 'o', markersize = 8, linestyle = '-', label = 'Weekly Mean Resample')
ax.set_ylabel('Solar Production (GWh)')
ax.legend()
plt.show()
```

```python
# ensure that only years with 359 valid days of data are summed
dopsd_annual = dopsd_daily[data_columns].resample('Y').sum(min_count = 359)

# set index of resampled dataframe and astype(ép kiểu dữ liệu)
dopsd_annual.index = dopsd_annual.index.year.astype(str)
dopsd_annual.index.name = 'Year'


# create a new column
dopsd_annual['Wind + Solar / Consumption'] = dopsd_annual['Wind+Solar'] / dopsd_annual['Consumption']

dopsd_annual.tail(3)
```

```python
dopsd_annual.head()
```

```python
# plot from 2012 onwards
ax = dopsd_annual.loc['2012':, 'Wind + Solar / Consumption'].plot.bar(color = 'C1')
ax.set_ylabel('Fraction')
ax.set_ylim(0, 0.3)
ax.set_title('Wind + Solar share of annual electricity Consumption')
plt.xticks(rotation = 0)
plt.show()
```

### 7. Rolling **windows**

```python
# compute the centered 14-day rolling mean
dopsd_14d = dopsd_daily[data_columns].rolling(14, center = True).mean()
dopsd_14d.head(10)
```

### 8. Trend

```python
dopsd_365 = dopsd_daily[data_columns].rolling(window = 365, center = True, min_periods = 360).mean()
dopsd_365.head(10)
```

```python
# create a figure with subplots
fig, ax = plt.subplots()
ax.plot(dopsd_daily['Consumption'], marker = '.', markersize = 2, color = 'C1',linestyle = None, label = 'Daily')
ax.plot(dopsd_14d['Consumption'], linewidth = 2, label = '14d Rolling Mean')
ax.plot(dopsd_365['Consumption'], color = 'C2', linestyle = None, linewidth = 3, label = '365d Rolling Mean')
ax.set_ylabel('Consumption (GWh)')
ax.set_xlabel('Year')
ax.set_title('Trends in Electricity Consumption')
ax.legend()
plt.show()
```

```python
import matplotlib.dates as mdates

fig, ax = plt.subplots()
for nm in ['Wind', 'Solar', 'Wind+Solar']:
  ax.plot(dopsd_365[nm], label = nm)
  ax.xaxis.set_major_locator(mdates.YearLocator())
  ax.set_ylim(0,400)
  ax.set_ylabel('Production (GWh)')
  ax.set_xlabel('Year')
  ax.set_title('Trends in Electricity Production')
  ax.legend()
```
