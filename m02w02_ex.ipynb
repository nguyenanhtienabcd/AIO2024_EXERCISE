### Compute vector length

```python
def compute_vector_length(vector):
    if isinstance(vector, np.ndarray) == False:
        vector = np.array(vector)
    length = np.linalg.norm(vector)
    return length
```

```python
vector = np.array([-2, 4, 9, 21])
result = compute_vector_length([ vector ])
print(round(result ,2))
```

### Vector dot Vector

```python
def compute_dot_product(vector_a, vector_b):
    if isinstance(vector_a, np.ndarray) == False:
        vector_a = np.array(vector_a)
    if isinstance(vector_b, np.ndarray) == False:
        vector_b = np.array(vector_b)

    return np.dot(vector_a, vector_b)
```

```python
v1 = np.array([0 , 1, -1, 2])
v2 = np.array([2 , 5, 1, 0])
result = compute_dot_product(v1 , v2)
print(round(result,2))
```

### Inverse matrix

```python
def compute_inverse_matrix(matrix):
    if isinstance(matrix, np.ndarray) == False:
        matrix = np.array(matrix)
    det = np.linalg.det(matrix)
    if det == 0:
        print(" determinant = 0, need to check the matrix ")
        return None
    else:
        return np.linalg.inv(matrix)
```

```python
m1 = np.array([[ -2 , 6], [8, -4]])
result = compute_inverse_matrix(m1)
print(result)
```

## Ex2: Eigenvector và eigenvalues



*  
$
\mathbf{A}\in\mathbb{R}^{n\times n}
$
,
$
\mathbf{I}\text{(identity matrix)}\in\mathbb{R}^{n\times n}
$
,
$
\mathbf{v}\in\mathbb{R}^{n\times n}
$

*   
$
\text{Eigenvalues } (λ): det(\mathbf{A} - λ\mathbf{I}) = 0
$

*  
$
\text{Eigenvectors } (\mathbf{v}): \mathbf{Av}=λ\mathbf{v}
\Leftrightarrow (\mathbf{A} - λ)\mathbf{v} = 0
$

*  
$
\text{Normalize vector}: \Large\frac{\mathbf{v}}{\|\mathbf{v}\|}\small,\text{ }\large v_i\small = \Large\frac{v_i}{\sqrt{\sum_{i=1}^n(v_i^2)}}
$



```python
import numpy as np

def compute_eigenvalues_eigenvectors(matrix):
    if isinstance(matrix, np.ndarray) == False:
        matrix = np.array(matrix)
    eigenvalue, eigenvector = np.linalg.eig(matrix)
    return eigenvalue, eigenvector


def norm_eigenvctor(eigenvector):
    length = np.linalg.norm(eigenvector)
    return eigenvector/length
```

```python
matrix = np.array([[0.9 , 0.2], [0.1 , 0.8]])
eigenvalues, eigenvectors = compute_eigenvalues_eigenvectors(matrix)
print(eigenvectors)
```

## Ex3: Cosine Similarity

*  
$
\text{Data(vector }
\mathbf{x,y}\text{): }
\mathbf{x} = \text{\{} x_1, x_2...x_n \text{\}, }
\mathbf{y} = \text{\{} y_1, y_2...y_n \text{\}}
$  
*  
$
\text{Cosine Similarity: cs(}\mathbf{x,y} \text{)} =
\frac{\Large \mathbf{x.y}}{\Large \|\mathbf{x} \|.\|\mathbf{y} \|}
=
\frac{\Large \sum_1^n(x_i y_i)}{\Large \sqrt{\sum_1^n x_i^2} \sqrt{\sum_1^n y_i^2}}
$


```python
import numpy as np
def compute_cosine(matrix_a, matrix_b):
    if isinstance(matrix_a, np.ndarray) == False:
        matrix_a = np.array(matrix_a)
    if isinstance(matrix_b, np.ndarray) == False:
        matrix_b = np.array(matrix_b)
    matrix_a = matrix_a.flatten()
    matrix_b = matrix_b.flatten()
    numerator = np.dot(matrix_a, matrix_b)
    denominator = np.linalg.norm(matrix_a) * np.linalg.norm(matrix_b)
    return numerator/denominator
```

```python
x = np.array([1 , 2, 3, 4])
y = np.array([1 , 0, 3, 0])
result = compute_cosine(x,y)
print(round (result, 3))
```

## Ex4: Background subtraction

### import ảnh, đổi hệ màu, thay đổi kích **thước**

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# đưa ảnh vào, đổi hệ màu, thay đổi kích thước của ảnh
bg = cv2.imread('/content/bacground.jpg',1)
bg = cv2.cvtColor(bg, cv2.COLOR_BGR2RGB)
bg = cv2.resize(bg,(1920, 2560))
fg = cv2.imread('/content/forceground.jpg',1)
fg = cv2.cvtColor(fg, cv2.COLOR_BGR2RGB)
fg = cv2.resize(fg,(1920, 2560))
obj = cv2.imread('/content/object.jpg',1)
obj = cv2.cvtColor(obj, cv2.COLOR_BGR2RGB)
obj = cv2.resize(obj,(1920, 2560))

# hiển thị các bức ảnh
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].set_title('Background')
ax[1].set_title('Forceground')
ax[2].set_title('Object')
ax[0].imshow(bg)
ax[1].imshow(fg)
ax[2].imshow(obj)
plt.show()
```

### tính toán sự khác nhau giữa hai ảnh nguồn

```python
# lấy hai ảnh có trừ cho nhau (ảnh có vật, và ảnh background ban đầu)
diff_img = cv2.absdiff(fg,obj)

# tạo thành ảnh xám (1 channel)
difference_gray = np.apply_along_axis(np.mean,  axis = -1, arr = diff_img)
plt.imshow(difference_gray)
plt.show()
print(difference_gray.shape)
```

### Tạo ảnh nhị phân

```python
# chuyển đổi nhị phân, những giá trị nào >255 sẽ chuyển thành 255, ngược lại là 0
binary_threshold = np.where(difference_gray > 55, 255, 0)
#binary_threshold = cv2.threshold(difference_gray, 55, 255, cv2.THRESH_BINARY)[1]

#----chuyển ảnh xám 1 kênh màu, thành ảnh xám 3 kênh màu------
# (gray_image,) là một tuple
# => (gray_image,)*3 = (gray_image, gray_image, gray_image)
# stack: xếp chồng 3 mảng theo trục axis = 2

gray_image_3_channel = np.stack((binary_threshold,)*3, axis=-1)

# nếu không dùng tuple mà để là "gray_image * 3"
# thì đây sẽ là ma trận nhân với hệ số 3 => sai mục đích
#---------------------------------------

plt.imshow(binary_threshold)
plt.show()
```

### chuyển đổi ảnh nhị phân thành ảnh màu

```python
output=np.where(gray_image_3_channel==0,bg,obj)
plt.imshow(output)
plt.show()
#---- tạo một ảnh mới trong folder____
output = cv2.cvtColor(output, cv2.COLOR_RGB2BGR)
cv2.imwrite('output.jpg',output)
```
